## 1. When do we go for Chi-square test?

The Chi-square test is used when we want to examine the relationship between two **categorical variables**.

It helps answer questions like:
- Is there an association between a person's education level and their job type?
- Does gender influence purchase decisions?

In your Kaggle notebook, the variables **Gender** and **Bought** are both categorical, making them suitable for a Chi-square test.

## 2. Do we use Chi-square test for independence? Explain how.

Yes, the Chi-square test is commonly used to test for **independence between two categorical variables**.

### How it works:
1. A contingency table is created using the two variables.
2. The Chi-square test calculates the difference between the **observed** and **expected frequencies** in each cell.
3. If the difference is large enough (statistically significant), we conclude that the variables are **not independent**.
