### 3. **What role does ANOVA play in data analytics?**

ANOVA helps in:

*   Comparing **multiple group means** at once without performing many t-tests.
    
*   Identifying **important factors** affecting a continuous variable.
    
*   Reducing dimensionality by focusing on **significant variables**.
    
*   Supporting **feature selection** in machine learning pipelines.
    

In short, ANOVA helps you figure out **which categorical variables significantly influence a numerical outcome**.

### 4. **When to use One-Way ANOVA? Explain with real-world case.**

Use **one-way ANOVA** when:

*   You have **one categorical independent variable** with 2 or more groups.
    
*   You want to see if it significantly affects a **numerical dependent variable**.
    

#### **Example:**

You want to know whether **students' performance** is influenced by the **type of teaching method** used:

*   Groups: Traditional, Online, Hybrid
    
*   Dependent variable: Final Exam Score
    

You would use one-way ANOVA to test:

> Do exam scores differ significantly across the three teaching methods?

### 5. **When to use Two-Way ANOVA? Explain with real-world case.**

Use **two-way ANOVA** when:

*   You have **two categorical independent variables**.
    
*   You want to see their **individual effects** and **interaction effect** on a numerical variable.
    

#### **Example:**

Suppose you're analyzing **crop yield** based on:

*   Fertilizer type (A, B, C)
    
*   Irrigation level (Low, Medium, High)
    

Here:

*   Two factors: Fertilizer and Irrigation
    
*   Dependent variable: Crop yield
    

Two-way ANOVA can tell you:

*   Does fertilizer type affect yield?
    
*   Does irrigation level affect yield?
    
*   Is there an interaction between fertilizer and irrigation?
