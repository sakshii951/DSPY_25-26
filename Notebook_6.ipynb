# Dataset Analysis and EDA

## 1. Description of Datasets

### IRIS Dataset
- **Number of features:** 4  
  - Sepal Length  
  - Sepal Width  
  - Petal Length  
  - Petal Width  
- **Categories:**
  - All 4 features are numerical (continuous)
  - Target variable is categorical (nominal) with 3 classes:
    - Setosa
    - Versicolor
    - Virginica
- **Number of tuples (rows):** 150  
- **Missing values:** No missing values  

---

### Air Quality Dataset
- **Number of features:** Around 15 (varies by source)  
  - Examples: CO(GT), NO2(GT), NOx(GT), Temperature, RH (Humidity), etc.
- **Categories:**
  - Most features are numerical (continuous)
  - Date and Time columns are text or datetime
- **Number of tuples:** ~9,358 (for one year of data)  
- **Missing values:** Yes  
  - Many values are marked as `-200` or `NaN`

---

### Wine Quality Dataset
- **Number of features:** 11 + 1 (target)
  - Examples: pH, alcohol, citric acid, fixed acidity, etc.
- **Categories:**
  - All 11 features are numerical (continuous)
  - Target variable (quality) is **ordinal** (scores from 0 to 10)
- **Number of tuples:** ~4,898 (combined red and white wine datasets)  
- **Missing values:** No missing values (in most versions)

## 2. Non-Graphical EDA in IRIS Dataset

Non-Graphical Exploratory Data Analysis (EDA) involves analyzing data without visualizations.

Common methods:

- `.head()` / `.tail()`  
  - Preview the first and last few rows of the dataset  

- `.describe()`  
  - Provides summary statistics:
    - Mean  
    - Standard deviation  
    - Minimum & maximum  
    - Quartiles  

- `.info()`  
  - Displays:
    - Data types  
    - Non-null counts  
    - Memory usage  

- `.value_counts()`  
  - Counts occurrences of each class in the target variable  

- `.corr()`  
  - Shows correlation between numerical features  

**Purpose:**  
Helps understand data distribution, central tendency, spread, and structure without using graphs.

## 4. Non-Graphical EDA in Wine Quality Dataset

Yes, non-graphical EDA can be performed on the Wine Quality dataset.

### Methods:

- `.describe()`  
  - Summarizes distribution of features like alcohol, acidity, etc.

- `.groupby("quality").mean()`  
  - Compares average feature values across quality scores  

- `.corr()`  
  - Identifies relationships between variables  
  - Example: correlation between alcohol and quality  

- `.quantile()`  
  - Helps detect outliers  

- `.skew()` and `.kurtosis()`  
  - Measure data distribution shape  

### Insights without graphs:
- Influence of features on wine quality  
- Feature variability  
- Detection of anomalies or unusual values
