## 1. One-Tailed and Two-Tailed T-Test

### Two-Tailed T-Test
- Tests whether the sample mean is significantly different from the population mean in **either direction** (greater or smaller)  
- Used when direction does not matter  

**Example:**
- H₀: μ = 19  
- H₁: μ ≠ 19  

---

### One-Tailed T-Test
- Tests whether the sample mean is significantly **greater than or less than** the population mean  
- Direction must be decided before testing  

**Example (Right-Tailed):**
- H₀: μ ≤ 19  
- H₁: μ > 19  

---

## 2. Type I and Type II Errors

### Type I Error (False Positive)
- Rejecting H₀ when it is actually true  

**Example:**  
You conclude average age ≠ 19, but it is actually 19  

---

### Type II Error (False Negative)
- Failing to reject H₀ when it is false  

**Example:**  
You accept average age = 19, but it is actually not 19  

---

## 3. Power of a Statistical Test

- **Power** = Probability of correctly rejecting a false null hypothesis  

**Formula:**

Power = 1 - β


- Where β = Probability of Type II Error  

**Interpretation:**
- High power → Better at detecting real effects  
- Low power → May miss true differences  

---
